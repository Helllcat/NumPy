```python
import numpy as np #numerical python
#array same colection of data with same datatype(homogenious)
a=np.array([1,2,3.1,4])
print(a)
```

```python
a=np.array([1,2,3.4,4,'a'])
a
```

```python
#slicing  foward and backward slicing
b=np.array([1,2,3,4,5])
print(b[0:2])
print(b[-1:-3:-1])
print(b[3:])
print(b[-4::-1])
```

```python
#dimension
import numpy as np
a=np.array([0,1,2,3,4,5])
c=np.array([[0,8],[1,2]])
print(a.ndim) #dimensiopn=1 vector
print(c.ndim)
```

```python

#create a 2x2 matrix 2,3,45
z=np.array([[2,3],[4,5]])
print(z)
print(z.shape)
print(z.ndim)
print(z.size)# size used for find total number of elements in amatrix
```

```python
c=np.array([[2,3,4],[4,5,5]])
print(c)
print(c.size)
print(c.ndim)
```
